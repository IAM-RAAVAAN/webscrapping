## For flipkart product prices 

```
import requests 
import re
import pandas as pd

s=requests.get("https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off").text
s = BeautifulSoup(s, 'lxml')
price=s.find_all('div',attrs={'class':"_30jeq3 _1_WHN1"})
prices=[]
names=[]
rattings=[]
links=[]
op=[]
e=[]
offer=[]
name=s.find_all('div',attrs={'class':"_4rR01T"})
ratting=s.find_all('div',attrs={'class':"gUuXy-"})
origi_pr=s.find_all('div',attrs={'class':"_3I9_wc _27UcVY"})

emi=s.find_all('div',attrs={'class':"_18hQoS"})

of=s.find_all('div',attrs={'class':"_2ZdXDB"})
z=zip(name,price,ratting,origi_pr,emi,of)
for n,p,r,ori,emi,of in z:
    names.append(n.text)
    prices.append(p.text)
    rattings.append(r.text)
    op.append(ori.text)
    e.append(emi.text)
    offer.append(of.text)
d={
    'name':names,
    'price':prices,
    'rattings':rattings,
    'original price':op,
    #'emi':e,
    'offers':offer
}
df=pd.DataFrame(d)
print(df)
df.to_csv('products.csv', index=False, encoding='utf-8')
```

##getting product prices for amazon

```
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

def url_maker(pro):
    url='https://www.amazon.in/s?k={}&ref=nb_sb_noss'
         
    return url.format(pro)


url='https://www.amazon.in/'

w=url_maker('mobile')
wd.get('https://www.amazon.in/')
soup=BeautifulSoup(wd.page_source,'html.parser')
print(soup)

result=soup.find_all('div',{"class":"s-include-content-margin s-border-bottom s-latency-cf-section"})
print(len(result))
lin_l=[]
names=[]
pri=[]
for page in range(1,4):
    wd.get(w)
    soup=BeautifulSoup(wd.page_source,'html.parser')
    result=soup.find_all('div',{"class":"s-include-content-margin s-border-bottom s-latency-cf-section"})
    for pro in result:
        pro_t=pro.h2.a.text
        names.append(pro_t)
        lin=pro.h2.a['href']
        lin_l.append(url+lin)
        ori_str=url+lin
        k=pro.find('span',{'class':"a-price-whole"})
        print(k.text)
        pri.append(k.text)
    s='https://www.amazon.in/s?k=mobile&page={}&qid=1612033571&ref=sr_pg_{}'
    w=s.format(page,page)
df={
    'links':lin_l,
    'names':names,
    'prizes':pri
}
df=pd.DataFrame(df)
print(df.names)
df.to_csv('amzon_data.csv', encoding='utf-8')
```
